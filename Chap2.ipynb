###雞兔同籠，若共有35個頭，94隻腳，請問有幾隻雞?幾隻兔?

```python
for chicken in range(0,36):
    rabbit=35-chicken #雞有幾顆頭，兔子就是剩下的頭
    if (2*chicken+4*rabbit) == 94: #兩隻腳和四隻腳的總和
        print("雞有",chicken,"隻,兔子有",rabbit,"隻")
        break
else:
    print("無解")
```

#while迴圈
先條件後規則

```python
sum=0
count=1
while count <=10:
    sum += count
    count += 1
print(sum)
```
