## 型態轉換

```python
x=12
print(type(x))

y=float(x)+0.5 #強制型別轉換

print(type(y))
print(y)
print(int(y))
print(int("12"))
print(float("12.5")) #不能用int
print(eval("5.5*2")) #自動算式
```

# print的三種方式


1.   透過,區隔要列印的變數 eg: print("This", "is", "a", "book")
2.   利用字串的str.format() (新式)
3.   %格式化输出字符串 (舊式)


```python
print("This"+"is"+"a"+"book") #字串相加
print("This","is","a","book") #字串物件
print()
print("This")#跟字串物件一樣
print("is")
print("a")
print("book")
```

```python
print("This {} a book".format("is")) #.format("1","2")需用{}指定放
print()
x=123456.789 #浮點數
print("{:.2f}".format(x))
   #指定小數點後兩位的浮點數
print("{:,.2f}".format(x))
   #百位上加,
```

```python
print("This %-10s a book %10s" % ("is","!")) #s=字串 d=數字 ；正數靠右,負數靠左 ；字串占10格，靠右2格
print()
print("%7.3f" % (687.4333)) # %7.=總共7位數，包含小數點
```
